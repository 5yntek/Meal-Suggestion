```python
from IPython.display import display
from PIL import Image

im = Image.open(r'.\sample_image_v1\banana.png') # Can be many different formats.
pix = im.load()
display(im)
```

```python
width, height = im.size
for x in range(width):
    for y in range(height):
        pixel = pix[x, y]
        if not pixel:
            print(pixel)
```
